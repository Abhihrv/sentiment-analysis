```python
def connect(username,timestamp, text, followers_count, friends_count, lang, place, location):
    """
    connect to MySQL database and insert twitter data
    """
    try:
        con = mysql.connector.connect(host = 'localhost',
        database='twitterdb', user='root', password = 'Password', charset = 'utf8')
    
        if con.is_connected():
            """
            Insert twitter data
            """
            cursor = con.cursor()
            # twitter, golf
            query = "INSERT INTO tweet_table (username,timestamp, text, followers_count, friends_count, lang, place, location) VALUES (%s, %s, %s, %s, %s, %s,%s, %s)"
            cursor.execute(query, (username,timestamp, text, followers_count, friends_count, lang, place, location))
            con.commit()

    except Error as e:
        print(e)

    cursor.close()
    con.close()

    return
```

```python

class MyStreamListener(tweepy.StreamListener):
        def on_connect(self):
            print("You are connected to the Twitter API")
            
        def on_data(self, data):
            
            try:
                raw_data = json.loads(data)
                if 'text' in raw_data:
                    print(raw_data['text'])
                    username = raw_data['user']['screen_name']
                    created_at = parser.parse(raw_data['created_at'])
                    tweet = raw_data['text']
                    followerCount=raw_data['user']['followers_count']
                    friendsCount=raw_data['user']['friends_count']
                    lang=raw_data['lang']
                    if raw_data['place'] is not None:
                        place = raw_data['place']['country']
                        print(place)
                    else:
                        place="NA"
                    location = raw_data['user']['location']
                    connect(username,created_at, tweet, followerCount, friendsCount, lang, place, location)
                print("Tweet colleted at: {} ".format(str(created_at)))
                    
                    
            except BaseException as e:
                print ('failed ondata'+str(e))
                time.sleep(5)
        def on_error(self, status):
            print (status)
```

```python
myStreamListener = MyStreamListener()
myStream = tweepy.Stream(auth = api.auth, listener=myStreamListener)
myStream.filter(track=['TwitchSupport'])
```
